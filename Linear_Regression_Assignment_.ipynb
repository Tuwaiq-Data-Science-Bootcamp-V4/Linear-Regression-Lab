# Our Goal: to predict the price given a set of attributes.

## Load the Data

```python
from sklearn import datasets
boston = datasets.load_boston()
```

## Q7: Split the dataset into Train and Test sets with test_size=30% and random_state=23.

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
x=df[['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX',
       'PTRATIO', 'B', 'LSTAT']]
y=df['Price']
```

```python
x_train, x_test ,y_train ,y_test=train_test_split(x,y,test_size=0.3,random_state=23)
```

## Q8: Build a Linear Regression Model.

```python
lr=LinearRegression()
```

## Q9: Train the Model.

```python
lr.fit(x_train,y_train)
```

## Q10: Evaluate the model. 
- print intercept and coefficients.
- compare between predictions and real values, then visualize them.
- Draw Residual Histogram.

```python
print(lr.intercept_) 
```

```python
coeff_df = pd.DataFrame(lr.coef_,x.columns,columns=['Coefficient'])
coeff_df
```

```python
pre=lr.predict(x_test)
pre
```

```python
RealV = np.array(y_test)
RealV
```

```python
plt.scatter(RealV,pre); 
```

## Residual Histogram

```python
sns.distplot((y_test-pre),bins=50);
```

## Q11: Use evaluation metrics MAE, MSE, RMSE and R^2.

```python
from sklearn import metrics
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, pre))
print('MSE:', metrics.mean_squared_error(y_test, pre))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, pre)))
```

```python
print('R^2:', metrics.r2_score(y_test, pre))
```
