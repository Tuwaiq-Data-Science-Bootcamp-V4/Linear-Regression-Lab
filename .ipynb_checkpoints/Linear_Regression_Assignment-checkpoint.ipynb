# Our Goal: to predict the price given a set of attributes.

## Load the Data

```python
from sklearn import datasets
boston = datasets.load_boston()
```

```python
X_boston,y_boston = boston.data, boston.target
print('Shape of data:', X_boston.shape, y_boston.shape)
```

## EDA (Exploratory Data Analysis)

## Q1: Create a dataframe and Save that dataset inside it.

```python
boston_df = pd.DataFrame(boston.data)
boston_df
```

## Q2: Print the head rows of the dataframe.

```python
boston_df.set_axis(['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO','B', 'LSTAT'], axis= 'columns', inplace= True)
```

```python
boston_df.head()
```

```python
boston_df['Price'] = y_boston.data
```

```python
boston_df.head()
```

## Q3: Use histogram to show the distribution of House Prices.

```python
plt.hist(boston_df['Price'])
plt.xlabel('Price')
plt.ylabel('count');
```

## Q4: Use a heatmap to show the correlation between features and the target labels.

```python
corrBoston = boston_df.corr()
```

```python
plt.figure(figsize=(12,7))
sns.heatmap(corrBoston, annot= True);
```

## Q5: Use a lmplot to draw the relations between price and LSTAT.

```python
plt.figure(figsize=(12,7))
sns.lmplot(boston_df, x = 'Price', y = 'LSTAT');
```

## Q6: Use a lmplot to draw the relations between price and RM.

```python
sns.lmplot(boston_df, x = 'Price', y = 'RM', aspect=2)
```

```python
boston_df.columns
```

## Q7: Split the dataset into Train and Test sets with test_size=30% and random_state=23.

```python
X = boston_df[['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX',
       'PTRATIO', 'B', 'LSTAT']]
y = boston_df['Price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=23)
```

## Q8: Build a Linear Regression Model.

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

## Q9: Train the Model.

```python
lm.fit(X_train,y_train)
```

## Q10: Evaluate the model. 
- print intercept and coefficients.
- compare between predictions and real values, then visualize them.
- Draw Residual Histogram.

```python
print(lm.intercept_)
```

```python
coeff_df = pd.DataFrame(lm.coef_,X.columns,columns=['Coefficient'])
coeff_df
```

```python
predictions = lm.predict(X_test)
predictions
```

```python
realValues = np.array(y_test)
realValues
```

```python
plt.scatter(predictions, y_test);
```

## Residual Histogram

```python
sns.distplot((y_test-predictions),bins=50);
```

## Q11: Use evaluation metrics MAE, MSE, RMSE and R^2.

```python
from sklearn import metrics
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test, predictions)
```
